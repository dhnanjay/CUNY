<h3>Geographical neighbors of India</h3>
<p>For this assignment, I would like to show all the geographical neighbors of India in graphical form</p>

```python
# Get Graph Instance
countries = nx.Graph()

# Create Nodes
countries.add_nodes_from(["India", "Pakistan", "China", "Nepal", "Bangladesh", "SriLanka", "Afghanistan", "Myanmar", "Bhutan"])

# Create Edges
countries.add_edges_from([("India","Pakistan"), ("India", "China"),("India","Bangladesh"), ("India","SriLanka"),("Pakistan","Afghanistan"),
                         ("Nepal", "China",), ("China", "Pakistan"), ("India", "Nepal"), ("India", "Myanmar"), ("Myanmar", "Bangladesh"),
                         ("India", "Bhutan"), ("Bhutan", "China")])
# Display Graph
nx.draw(countries, with_labels=True)
```

<h3> Analyzing Network </h3>

```python
# Size
print("Total Countries :",len(countries))
```

```python
# Nodes
countries.node
```

```python
# Edges
countries.edges
```

```python
# List Degree
import pandas
degrees = pandas.DataFrame(list(countries.degree()),columns=("country","degree")).set_index("country")
degrees.sort_values(by='degree', ascending=False).head(10)
```
